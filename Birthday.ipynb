# Birthday Problem
The **Birthday Problem** refers to the counterintuitive probability that, in a group of randomly chosen people, at least two of them will share the same birthday. Despite there being 365 possible birthdays, it only takes a group of **23** people for there to be a greater than 50% chance that two people share a birthday. The probability increases rapidly with the size of the group, approaching near certainty with just 70 people. This surprising result arises from the number of possible pairs increasing much faster than the number of people.

```python
import numpy as np
import matplotlib.pyplot as plt


def get_probab(n):
  # Calculations
  events = 365
  list_of_probab = []
  output_probab = 1
  for i in range(1, n+1):
    probab = output_probab * (events / 365)
    list_of_probab.append(probab)
    output_probab = probab
    events -= 1

  # Visualizations
  x_values = np.arange(1, n+1, dtype=np.int64)
  y_values = list_of_probab

  fig, ax = plt.subplots()
  ax.scatter(x_values, y_values)

  plt.xlabel("Number Of People")
  plt.ylabel("Probability Of Two People Sharing A Birthday")

  plt.show()
```

```python
get_probab(100)
```
